````python
def scrape_book(book_url):
  book_info = r.get(book_url).content
  book_soup = BeautifulSoup(markup=book_info, parser="html.parser")
  book_data = {}
  # getting name
  name = book_soup.find(name="h1").getText()
  book_data['name'] = name
  # getting other/table data
  book_table_data = book_soup.find_all(name="tr")
  for row in book_table_data:
    key = row.find(name="th").getText()
    value = row.find(name="td").getText()
    book_data[key] = value
  # let's also keep the url of book in final result
  book_data['url'] = book_url
  return book_data
# let's test this
scrape_book(book_url)
````

## **Scraping Book Page**

````python
def scrap_book(book_url):
  book_info = r.get(book_url).content
  book_soup = BeautifulSoup(markup = book_info, parser = 'html.parser')

  book_data = {}

  # getting book name
  name = book_soup.find(name='h1').getText()
  book_data['Name'] = name

  # getting other/table data
  book_table_data = book_soup.find_all(name = 'tr')
  for i in book_table_data:
    key = i.find(name = 'th').getText()
    value = i.find(name = 'td').getText()
    book_data[key] = value

  # let's also keep the url of book in final result
  book_data['url'] = book_url
  return book_data
scrap_book(book_url)
````
